```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Fortnite_Statistics.csv')  # Replace 'fortnite_stats.csv' with your dataset file path
```

```python
# Convert 'Accuracy' to a string and then to a numerical format (float)
df['Accuracy'] = df['Accuracy'].astype(str).str.rstrip('%').astype(float)
```

```python
# Convert 'Mental State' to integer values (1 for 'high', 0 for 'sober')
df['Mental State'] = df['Mental State'].map({'high': 1, 'sober': 0})
```

```python
# Drop the 'Date' and 'Time of Day' columns
df = df.drop(['Date', 'Time of Day'], axis=1)
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info(),'\n')
```

```python
df.head(1)
```

```python
# Distribution of 'Placed' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Placed'], bins=20, kde=True)
plt.title('Distribution of Placed')
plt.xlabel('Placed')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap for numerical columns
plt.figure(figsize=(10, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a scatter plot for 'Damage Taken' vs. 'Materials Used'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Damage Taken', y='Materials Used', hue='Placed')
plt.title('Damage Taken vs. Materials Used')
plt.xlabel('Damage Taken')
plt.ylabel('Materials Used')
plt.legend(title='Placed')
plt.show()
```
